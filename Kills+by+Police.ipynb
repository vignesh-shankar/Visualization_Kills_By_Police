```python
#importing required packages
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
import seaborn as sns
sns.set(color_codes=True)
```

```python
ds = pd.read_csv("C:\Users\Vignesh\Documents\KillsByPolice.csv")
```

```python
#Plotting the number of kills by year
#We can ibserve that there has been a decrease in the killings in 2016 as compared to 2015
plt.figure(figsize=(4,6))
sns.countplot(data = ds, x = "year")
```

```python
#Extending on the previous visualiztion by adding another dimension, Race
#The number of whites killed is significantly high compared to other races. 

sns.countplot(data = ds, x = "year", hue = "raceethnicity")
```

```python
#People killed, distributed by age. 
#We can observe from the plot that the number of killings is high between the age 25 and 40
#Shocking fact is that children below the age of 10 have bene killed, which definitely should be a accident.
age = ds.age.replace(to_replace="Unknown", value = None)
age = age.convert_objects(convert_numeric=True)
age = age.dropna()
sns.distplot(age, kde= False, bins = 20, axlabel = "Age",color = "red")
```

```python
#Just a kde plot of the above depicted distribution plot
sns.kdeplot(age, color = "red", shade=True, cut = 1)
```
